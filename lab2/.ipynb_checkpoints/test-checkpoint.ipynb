```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from skimage.metrics import structural_similarity
import pandas as pd
```

```python
test_img = cv.imread('../lab1/standard_test_images/cameraman.tif', cv.IMREAD_GRAYSCALE)
test_img = cv.normalize(test_img, None, 0, 1, cv.NORM_MINMAX, cv.CV_32F)

plt.imshow(test_img, cmap='gray')
plt.axis('off')

plt.show()
```

```python
def salt_and_pepper(img, noise):
    black_points = np.zeros(img.shape)
    white_points = np.ones(img.shape)

    mask = np.random.choice([0, 1, 2], size=img.shape, p=[1-noise, noise/2, noise/2])

    stacked = np.stack([img, black_points, white_points], axis=0)

    out = stacked[mask.reshape(1, mask.shape[0], mask.shape[1]), np.arange(img.shape[0])[:, None], np.arange(img.shape[1])].reshape(img.shape)
    
    return out

def impulse_floating(img, noise):
    noise_matrix = np.random.rand(img.shape[0], img.shape[1])
    
    mask = np.random.choice([0, 1], size=img.shape, p=[1-noise, noise])

    stacked = np.stack([img, noise_matrix], axis=0)

    out = stacked[mask.reshape(1, mask.shape[0], mask.shape[1]), np.arange(img.shape[0])[:, None], np.arange(img.shape[1])].reshape(img.shape)
    
    return out

def gaussian_noise(img, std):
    noise = np.random.normal(0, std, img.shape)
    
    out = img + noise
    
    out[out > 1] = 1
    out[out < 0] = 0
    
    return out
```

```python
def avg_filter(img, kernel_size=(3,3)):
    kernel = np.ones(kernel_size) / (kernel_size[0] * kernel_size[1])
    
    out = cv.filter2D(img, -1, kernel)
    
    return out

def gaussian_filter(img, sigma, kernel_size):
    m, n = kernel_size
    m_half = m // 2
    n_half = n // 2

    gaussian_kernel = np.zeros((m, n), np.float32)

    for y in range(-m_half, m_half):
        for x in range(-n_half, n_half):
            normal = 1 / (2.0 * np.pi * sigma**2.0)
            exp_term = np.exp(-(x**2.0 + y**2.0) / (2.0 * sigma**2.0))
            gaussian_kernel[y+m_half, x+n_half] = normal * exp_term

    return cv.filter2D(img, -1, gaussian_kernel)

def median_filter(img, filter_size):
    temp = []
    indexer = filter_size // 2
    result = []
    result = np.zeros((len(img),len(img[0])))
    for i in range(len(img)):
        for j in range(len(img[0])):
            for z in range(filter_size):
                if i + z - indexer < 0 or i + z - indexer > len(img) - 1:
                    for c in range(filter_size):
                        temp.append(0)
                else:
                    if j + z - indexer < 0 or j + indexer > len(img[0]) - 1:
                        temp.append(0)
                    else:
                        for k in range(filter_size):
                            temp.append(img[i + z - indexer][j + k - indexer])

            temp.sort()
            result[i][j] = temp[len(temp) // 2]
            temp = []

    return result
```

```python
def PSNR(original, compressed): 
    mse = np.mean((original - compressed) ** 2)
    if(mse == 0):  # MSE is zero means no noise is present in the signal.
                  # Therefore PSNR have no importance.
        return 100
    max_pixel = 255.0
    psnr = 20 * np.log10(max_pixel / np.sqrt(mse))
    return psnr

def SSIM(original, processed):
    return structural_similarity(original, processed, data_range=processed.max() - processed.min())
```

```python
salt_pepper_p = .1
salted_img = salt_and_pepper(test_img, salt_pepper_p)

plt.imshow(salted_img, cmap='gray')
plt.title(f'salt and pepper ({salt_pepper_p})')
plt.axis('off')

plt.show()
```

```python
impulse_floating_p = .1
impulse_img = impulse_floating(test_img, impulse_floating_p)

plt.imshow(impulse_img, cmap='gray')
plt.title(f'impulse floating ({impulse_floating_p})')
plt.axis('off')
plt.show()
```

```python
gaussian_p = .1
gaussian_img = gaussian_noise(test_img, gaussian_p)

plt.imshow(gaussian_img, cmap='gray')
plt.title(f'gaussian noise ({gaussian_p})')
plt.axis('off')
plt.show()
```

```python
kernel_size = (3,3)

plt.figure(figsize=(10,4))
plt.suptitle(f'avg filter ({kernel_size})')

plt.subplot(131)
plt.title('salted')
plt.imshow(avg_filter(salted_img, kernel_size), cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title('impulse floating')
plt.imshow(avg_filter(impulse_img, kernel_size), cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.title('gaussian')
plt.imshow(avg_filter(gaussian_img, kernel_size), cmap='gray')
plt.axis('off')

plt.show()
```

```python
kernel_size_gauss = (3,3)
sigma = 3

plt.figure(figsize=(10,4))
plt.suptitle(f'gaussian filter ({sigma}, {kernel_size_gauss})')

plt.subplot(131)
plt.title('salted')
plt.imshow(gaussian_filter(salted_img, sigma, kernel_size_gauss), cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title('impulse floating')
plt.imshow(gaussian_filter(impulse_img, sigma, kernel_size_gauss), cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.title('gaussian')
plt.imshow(gaussian_filter(gaussian_img, sigma, kernel_size_gauss), cmap='gray')
plt.axis('off')

plt.show()
```

```python
diameter = 7
sigma_i = 20
sigma_s = 20

plt.figure(figsize=(10,4))
plt.suptitle(f'bilateral filter ({diameter}, {sigma_i}, {sigma_s})')

plt.subplot(131)
plt.title('salted')
plt.imshow(cv.bilateralFilter(np.uint8(salted_img * 255), diameter, sigma_i, sigma_s), cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title('impulse floating')
plt.imshow(cv.bilateralFilter(np.uint8(impulse_img * 255), diameter, sigma_i, sigma_s), cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.title('gaussian')
plt.imshow(cv.bilateralFilter(np.uint8(gaussian_img * 255), diameter, sigma_i, sigma_s), cmap='gray')
plt.axis('off')

plt.show()
```

```python
median_filter_size = 3

plt.figure(figsize=(10,4))
plt.suptitle(f'median filter ({median_filter_size})')

plt.subplot(131)
plt.title('salted')
plt.imshow(median_filter(salted_img, median_filter_size), cmap='gray')
plt.axis('off')

plt.subplot(132)
plt.title('impulse floating')
plt.imshow(median_filter(impulse_img, median_filter_size), cmap='gray')
plt.axis('off')

plt.subplot(133)
plt.title('gaussian')
plt.imshow(median_filter(gaussian_img, median_filter_size), cmap='gray')
plt.axis('off')

plt.show()
```

```python
dic = [
    {'name': 'noise', 'lambda': lambda x: x},
    {'name': 'avg', 'lambda': lambda x: avg_filter(x, kernel_size)},
    {'name': 'gaussian', 'lambda': lambda x: gaussian_filter(x, sigma, kernel_size_gauss)},
    {'name': 'bilateral', 'lambda': lambda x: cv.bilateralFilter(np.uint8(x * 255), diameter, sigma_i, sigma_s)},
    {'name': 'median', 'lambda': lambda x: median_filter(x, median_filter_size)},
]

data = [{'name': method['name'], 'SSIM (impulse 1)': SSIM(test_img, method['lambda'](salted_img)), 'SSIM (impulse 2)': SSIM(test_img, method['lambda'](impulse_img)), 'SSIM (gauss)': SSIM(test_img, method['lambda'](gaussian_img)), 'PSNR (impulse 1)': PSNR(test_img, method['lambda'](salted_img)), 'PSNR (impulse 2)': PSNR(test_img, method['lambda'](impulse_img)), 'PSNR (gauss)': PSNR(test_img, method['lambda'](gaussian_img))} for method in dic]

pd.DataFrame(data)
```

```python
pd.DataFrame(data).to_csv('./out.csv', index=False, float_format='%.2f')
```
